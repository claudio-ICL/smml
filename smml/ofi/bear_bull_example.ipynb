```python
from smml import ofi
import numpy as np
import matplotlib
```

```python
ld = ofi.data.LobsterData(ofi.data.LDI_EG)
```

```python
df = ld.unique_time_ob(t0=None, t1=None, alpha=0.00025)
```

```python
df.loc[300000:400000][['mid_price', 'smooth_mid']].plot()
```

```python
idx_bear = df['bear_bull'] == -1
df_bull = df.copy()
df_bull.loc[idx_bear, :] = np.nan
df_bull.loc[300000:400000][['mid_price', 'smooth_mid']].plot()
```

```python
idx_bull = df['bear_bull'] == 1
df_bear = df.copy()
df_bear.loc[idx_bull, :] = np.nan
df_bear.loc[300000:400000][['mid_price', 'smooth_mid']].plot()
```

```python
samples = ofi.data._from_ob_to_volume_samples_dataframe(df, levels=3)
```
